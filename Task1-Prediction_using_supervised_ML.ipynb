# **TASK 1: Prediction Using Supervised ML**

In this task we are provided with a simple dataset of the number of hours studied by a student vs their score. We have to model this data using supervised ML technique linear regression and use that to make predictions on new data.

## IMPORTING LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## LOADING DATA

```python
# Loading the required data using read_csv from url
data = pd.read_csv("https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv")
print("Data imported successfully")
data.head()
```

```python
# A brief description of our data
data.info()
```

### Now I check for any null values. Also I check the shape of the dataframe

```python
# checking for null values
data.isnull()
```

```python
data.shape
```

### I plot the data on a scatter plot to get an overview of its distribution

```python
# Plotting the distribution of scores
data.plot(x='Hours', y='Scores', style='o')
plt.title('Hours vs Percentage')
plt.xlabel('Hours Studied')
plt.ylabel('Percentage Score')
plt.show()
```

## TRAINING THE MODEL

```python
#Distributing the data into attributes and labels
X = data.iloc[:,:-1].values
y = data.iloc[:, 1].values
```

```python
# splitting the data into testing and training data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y,random_state=0)
```

```python
# importing from library the model for linear regression
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
print("Model trained successfully")
```

```python
# plotting the regression line
line = model.coef_*X+model.intercept_
plt.scatter(X,y)
plt.plot(X,line)
plt.show()
```

## PREDICTIONS ON TESTING DATA

Now I make predictions by running the trained model on the testing data

```python
# the testing data and calculating the output on the testing data to store in y_pred
print(X_test)
y_pred = model.predict(X_test)
```

```python
# Comparing Actual vs Predicted
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df
```

## EVALUATING THE MODEL

Now I evaluate the accuracy of our predictions by using the absolute mean square error metric

```python
# importing the absolute mse measure and testing model accuracy
from sklearn.metrics import mean_absolute_error
mse = mean_absolute_error(y_test, y_pred)
print(mse)
```

## PREDICTION ON NEW DATA

```python
# You can also test with your own data
hours = 9.25
own_pred = model.predict([[hours]])
print("No of Hours = {}".format(hours))
print("Predicted Score = {}".format(own_pred[0]))
```
